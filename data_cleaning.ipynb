```python
#menghapus spasi berlebih
df['title'] = df['title'].str.strip().str.replace(r'\s+', ' ', regex=True)

df.head()
```

```python
#menghilangkan karakter khusus
df['title'] = df['title'].str.replace(r'[^\w\s]', '', regex=True)

#mengatasi kata yang berulang
df['title'] = df['title'].str.replace(r'\b(\w+)\s+\1\b', r'\1', regex=True)

#hapus value yang kosong
df = df[df['title'].notnull() & ~df['title'].str.contains('unknown|tidak ada', na=False)]
```

```python
#setelah dilihat terdapat bahasa cina pada title
#deteksi entri dengan karakter non-Latin
non_latin_titles = df[df['title'].str.contains(r'[^\x00-\x7F]', na=False)]

for idx, title in non_latin_titles['title'].items():
    print(f"Index {idx}: {title}")
```

```python
def clean_text(text):
    if isinstance(text, str):  # Pastikan hanya string yang diproses
        # Ganti '/' dengan 'atau' dan '&' dengan 'dan'
        text = text.replace('/', 'atau').replace('&', 'dan')
        # Hilangkan spasi di awal, akhir, dan spasi berlebih antar kata menggunakan re.sub
        text = re.sub(r'\s+', ' ', text.strip())
    return text

df = df.applymap(clean_text)
```

```python
import re
from nltk.corpus import wordnet
from nltk.stem import WordNetLemmatizer
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory
```

```python
df.head()
```

```python
# Inisialisasi stemmer Sastrawi untuk Bahasa Indonesia
factory = StemmerFactory()
stemmer_indo = factory.create_stemmer()
```

```python
# Inisialisasi lemmatizer NLTK untuk Bahasa Inggris
lemmatizer_en = WordNetLemmatizer()
```

```python
#hapus tanda baca, emotikon, dan hurus non latin.
def clean_text(text):
    text = re.sub(r'[^\w\s]', '', text)  # Hapus tanda baca
    text = re.sub(r'\d+', '', text)      # Hapus angka
    text = re.sub(r'[^\x00-\x7F]+', '', text)  # Hapus karakter non-ASCII
    return text.lower()
```

```python
# Lematisasi campuran bahasa
def lemmatize_mixed_language(text):
    words = text.split()
    lemmatized_words = []
    for word in words:
        # Lematisasi menggunakan NLTK untuk kata Bahasa Inggris
        if wordnet.synsets(word):
            lemmatized_word = lemmatizer_en.lemmatize(word.lower())
        else:
            # Lematisasi menggunakan Sastrawi untuk kata Bahasa Indonesia
            lemmatized_word = stemmer_indo.stem(word.lower())
        lemmatized_words.append(lemmatized_word)
    return ' '.join(lemmatized_words)
```

```python
df['cleaned_description'] = df['description'].apply(lambda x: lemmatize_mixed_language(clean_text(str(x))))
print(df[['description', 'cleaned_description']].head())
```

```python
# Menggabungkan stopword Bahasa Indonesia dan Inggris
stopwords_combined = stopwords_indo.union(stopwords_en)
```

```python
def remove_stopwords(text):
    words = word_tokenize(text)  # Tokenisasi kata
    filtered_words = [word for word in words if word not in stopwords_combined]
    return ' '.join(filtered_words)
```
